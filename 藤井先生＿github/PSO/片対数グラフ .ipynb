```python
N = 2

# xの値域は自動化しておく
min_field_list = [-5.0, -5.0, -5.0, -600.0, -10.0, -5.0]
max_field_list = [5.0, 5.0, 10.0, 600.0, 10.0, 5.0]
min_field = min_field_list[function_number - 1]
max_field = max_field_list[function_number - 1]
```

```python
def plot_single_log(function_number, x, y, title = "", title_fontsize = 24, xlabel = "", ylabel = "", x_fontsize = 18, y_fontsize = 18) : 
    f = Function(function_number)
    score_list = []
    for place in y : 
        score_list.append(f(place[0]))
    print(score_list)
    
    fig, ax = plt.subplots()
    ax.semilogy(x, score_list)
    ax.grid()
    plt.title(title, fontsize = title_fontsize)
    plt.xlabel(xlabel, fontsize = x_fontsize)
    plt.ylabel(ylabel, fontsize = y_fontsize)
    plt.show()
```
