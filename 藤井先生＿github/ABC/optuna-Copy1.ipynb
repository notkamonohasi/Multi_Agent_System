```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.animation as animation
from IPython.display import HTML
import optuna

from ABC import ABC

INF = 10000000000

%matplotlib inline
```

```python
def objective(trial):
    x = trial.suggest_uniform('x', -10, 10)
    score = (x - 2) ** 2
    print('x: %1.3f, score: %1.3f' % (x, score))
    return score

study = optuna.create_study()
study.optimize(objective, n_trials=100)
```

```python
# パラメータ
N = 2   # 次元
function_number = 2
limit_time = 1000   # 1秒
repeat_times = 3

# xの値域は自動化しておく
min_field_list = [-5.0, -5.0, -5.0, -600.0, -10.0, -5.0]
max_field_list = [5.0, 5.0, 10.0, 600.0, 10.0, 5.0]
min_field = min_field_list[function_number - 1]
max_field = max_field_list[function_number - 1]

def objective(trial) : 
    num_harvest_bee = trial.suggest_int("num_harvest_bee", 20, 100)
    num_follow_bee = trial.suggest_int("num_follow_bee", 20, 100)
    limit_harvest = trial.suggest_int("limit_harvest", 10, 100)
    max_turn = trial.suggest_int("max_turn", 100, 500)
    
    sum_score = 0
    for _ in range(repeat_times) : 
        abc = ABC(N, function_number, num_harvest_bee, num_follow_bee, min_field, max_field, limit_harvest, max_turn)
        if abc.calculation_time < limit_time :   # 時間制限を超えていない時
            sum_score += abc.gb_score
        else : 
            sum_score += INF
            
    return sum_score
```

```python
study = optuna.create_study()
study.optimize(objective, n_trials=100)
```

```python
print(study.best_params)
print(study.best_value)
print(study.best_trial)
```
